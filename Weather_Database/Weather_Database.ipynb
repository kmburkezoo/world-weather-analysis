```python
# 0. Import dependencies
import pandas as pd
import numpy as np
import requests
from citipy import citipy
from pprint import pprint

#keys
from config import g_key
from config import weather_api_key
```

```python
# 1. Create set of 2,000 random coordinates
lats = np.random.uniform(-90.000,90.000,size=2000)
lngs = np.random.uniform(-180.000,180.000,size=2000)
lat_lngs = zip(lats,lngs)

# Add coordinates to a list
coordinates = list(lat_lngs)
coordinates
```

```python
#2. Get nearest cities using citipy module
# create list to hold cities
cities = []

# identify nearest city for each set of coordinates
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0],coordinate[1]).city_name
    
    # add unique cities to cities list
    if city not in cities:
        cities.append(city)
        
# print city count to see how many we have
len(cities)
```
